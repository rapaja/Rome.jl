```julia
using Romeo
ω = 10 .^ range(-4, stop=4, length=1000);
```

```julia
G = (α, β) -> (im .* ω) .^ (-α -1im*β);
Gs = (α, β) -> 0.5 * ((im .* ω) .^ (-α -1im*β) + (im .* ω) .^ (-α +1im*β));
```

```julia
using Plots
```

```julia
plot(ω, NaN*ω)
for a in range(0, stop=1, step=0.5)
    for b in [1] #b in range(0, stop=1, step=0.5)
        plot!(ω, abs.(Gs(a,b)), xaxis=:log, yaxis=:log, label="α=$a, β=$b")
    end
end
plot!(ω, NaN*ω)
# plot(plots...)
```

```julia
plot(ω, NaN*ω)
for a in [1] # a in range(0, stop=1, step=0.5)
    for b in range(0, stop=1, step=0.5)
        an = angle.(Gs(a,b))
        Romeo.unwrap!(an)
        plot!(ω, an, xaxis=:log, label="α=$a, β=$b")
    end
end
plot!(ω, NaN*ω)
# plot(plots...)
```
